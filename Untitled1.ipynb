```python
import pygame
import sys

# Initialize Pygame
pygame.init()

# Set up the screen
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Bouncing Ball")

# Colors
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
RED = (255, 0, 0)

# Ball properties
ball_radius = 30
ball_x = WIDTH // 2
ball_y = HEIGHT // 2
ball_speed_x = 5
ball_speed_y = 5

# Main loop
running = True
while running:
    screen.fill(BLACK)

    # Event handling
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Update ball position
    ball_x += ball_speed_x
    ball_y += ball_speed_y

    # Bounce off walls
    if ball_x + ball_radius > WIDTH or ball_x - ball_radius < 0:
        ball_speed_x = -ball_speed_x
    if ball_y + ball_radius > HEIGHT or ball_y - ball_radius < 0:
        ball_speed_y = -ball_speed_y

    # Draw ball
    pygame.draw.circle(screen, RED, (ball_x, ball_y), ball_radius)

    # Update display
    pygame.display.flip()

    # Control frame rate
    pygame.time.Clock().tick(60)

# Quit Pygame
pygame.quit()
sys.exit()
```

```python
import pyautogui
import time

def track_mouse():
    try:
        while True:
            x, y = pyautogui.position()
            print(f"Mouse position: x={x}, y={y}")
            time.sleep(1)  # Adjust the delay as needed
    except KeyboardInterrupt:
        print("\nTracking stopped.")

if __name__ == "__main__":
    track_mouse()
```

```python
import random

def play_cricket():
    print("Welcome to Text-based Cricket Game!")
    print("You are batting against the computer bowler.")
    print("You have 6 balls to score as many runs as possible.")

    total_runs = 0
    remaining_balls = 6

    while remaining_balls > 0:
        input("Press Enter to bat...")
        player_runs = random.randint(0, 6)
        computer_runs = random.randint(0, 6)

        print(f"You: {player_runs} | Computer: {computer_runs}")

        if player_runs != computer_runs:
            print("You scored runs!")
            total_runs += player_runs
        else:
            print("You're out!")
            break

        remaining_balls -= 1

    print(f"Your total runs: {total_runs}")
    print("Game over!")

if __name__ == "__main__":
    play_cricket()
```
